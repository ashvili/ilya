## 2. Загрузка данных
Загрузите файл `explored_blocks.csv` в папку `temp`.

```python
import os
from google.colab import files

os.makedirs('temp', exist_ok=True)

print("Пожалуйста, загрузите файл 'explored_blocks.csv':")
uploaded = files.upload()

for filename in uploaded.keys():
    # Перемещаем загруженный файл в temp/
    os.rename(filename, os.path.join('temp', filename))
    print(f"Файл {filename} перемещен в temp/")
```

## 3. Настройка и запуск
Здесь можно изменить конфигурацию перед запуском.

```python
import json

config_path = 'config.json'

with open(config_path, 'r') as f:
    config = json.load(f)

# --- НАСТРОЙКИ ---
config['ml']['epochs'] = 30
config['ml']['device'] = 'cuda' # Используем GPU
config['ml']['blocks_file_path'] = './temp/explored_blocks.csv'
# -----------------

with open(config_path, 'w') as f:
    json.dump(config, f, indent=4)

print("Конфигурация готова. Запускаем обучение...")

!python -m ml
```
